```python
def CalEff(DF1,nm):
    evt=DF1.iloc[:,0]
    for i in range(1,10,2):
        print(nm[i])
        eff=DF1.iloc[:,i]
        coun=eff[0]
        for j in range(eff.shape[0]):
            if j==0:
                rat=(eff[0]/eff[0])*100
                print('Cut:', evt[j] ,rat)
            else:
                rat=(eff[j]/eff[j-1])*100
                print('Cut:', evt[j], '%0.02f'% rat)
        print('\n')
```

```python
CalEff(data2,names)
```

```python
def CalEff2(DF1,nm):
    evt=DF1.iloc[:,0]
    for i in range(1,10,2):
        print(nm[i])
        eff=DF1.iloc[:,i]
        coun=eff[0]
        for j in range(eff.shape[0]):
            if j==0:
                rat=(eff[0]/eff[0])*100
                print('Cut:', evt[j] ,rat)
            else:
                rat=(eff[j]/coun)*100
                coun+=eff[j]
                print('Cut:', evt[j], '%0.02f'% rat, 'Cumulative: %0.02f '% coun)
        print('\n')
```

```python
CalEff2(data2,names)
```
